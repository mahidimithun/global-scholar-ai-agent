```python
import os
from google.colab import userdata
from google import genai
from google.genai import types
from IPython.display import display, Markdown

# 1. Setup Client
os.environ["GEMINI_API_KEY"] = userdata.get('GEMINI_API_KEY')
client = genai.Client()

print("🎉 Agent is initialized and waiting for you...")
print("-" * 60)

# =====================================================================
# 📥 LIVE INTERACTIVE INPUT STEP
# =====================================================================
USER_INPUT = input("👉 Enter your relocation inquiry (e.g., My degree is CSE, show me cheap options...): ")

print("\n🕵️ Agent Status: Input received! Activating live search grounding...")
print("-" * 60)

# 2. Deep System Instructions
system_instruction = (
    "You are an expert Concierge Relocation Agent. Your goal is to help users find the absolute best "
    "and most affordable destinations based on their academic background and goals. You must prioritize "
    "factual integrity by utilizing your native Google Search tool to fetch live, real-time statistics on "
    "international student tuition fees, affordable universities for Computer Science/CSE, monthly rental costs, "
    "and state-specific cost of living indices in Australia.\n\n"
    "Your response MUST contain two specific sections:\n"
    "1. A cleanly structured Markdown comparison table summarizing the cheapest states and universities found.\n"
    "2. A chronological, personalized pre-departure and academic preparation checklist."
)

user_prompt = f"""
The user has provided the following profile and request:
"{USER_INPUT}"

Please execute real-time search grounding to find the most budget-friendly options matching their criteria.
Evaluate tuition fees, local rent trends, and general living affordability. Compile everything into a
structured Blueprint comparison table and an action-item checklist.
"""

# =====================================================================
# 🛠️ CORRECTED GOOGLE SEARCH TOOL INITIALIZATION
# =====================================================================
# Explicitly using the exact types expected by the Pydantic validator
grounding_tool = types.Tool(
    google_search=types.GoogleSearch()
)

# 3. Execute Agent
try:
    response = client.models.generate_content(
        model='gemini-2.5-flash',
        contents=user_prompt,
        config=types.GenerateContentConfig(
            system_instruction=system_instruction,
            tools=[grounding_tool], # Correct typed tool object passed here
            temperature=0.3
        )
    )

    # Display Results
    display(Markdown("# 🧭 PERSONALIZED LIFESTYLE & RELOCATION BLUEPRINT"))
    display(Markdown(response.text))

    # Grounding Logs
    if response.candidates[0].grounding_metadata.web_search_queries:
        print("\n" + "="*60)
        print("🔍 AGENT LOG: Google Search Queries Performed for Grounding:")
        for query in response.candidates[0].grounding_metadata.web_search_queries:
            print(f" -> {query}")
        print("="*60)

except Exception as e:
    print(f"An unexpected execution error occurred: {e}")
```
